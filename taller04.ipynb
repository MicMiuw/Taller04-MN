## Graficar la función cos(x) con X0= 0.

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-3, 3, 1000)
y = np.cos(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Gráfica de la función coseno')
plt.grid()
plt.show()
```

### SERIE DE TAYLOR ORDEN 0

```python
taylor_0 = np.ones_like(x)
plt.plot(x, y, label='cos(x)')
plt.plot(x, taylor_0, label='Taylor orden 0', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función coseno y su aproximación de Taylor')
plt.axvline(x=0, color='red', linestyle='--', label='Punto de expansión x=0')
plt.grid()
plt.show()
```

### SERIE DE TAYLOR ORDEN 2

```python
taylor_2 = 1- (x**2)/2 
plt.plot(x, y, label='cos(x)') 
plt.plot(x, taylor_2, label='Taylor orden 2', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.axvline(x=0, color='red', linestyle='--', label='Punto de expansión x=0')
plt.grid()
plt.legend()
plt.show()
```

### SERIE DE TAYLOR ORDEN 4

```python
taylor_4 = 1- (x**2)/2 + (x**4)/24
plt.plot(x, y, label='cos(x)')
plt.plot(x, taylor_0, label='Taylor orden 0', linestyle='--')
plt.plot(x, taylor_2, label='Taylor orden 2', linestyle='--')
plt.plot(x, taylor_4, label='Taylor orden 4', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función coseno y sus aproximaciones de Taylor')
plt.axvline(x=0, color='red', linestyle='--', label='Punto de expansión x=0')
plt.grid()
plt.legend()
plt.show()
```

### SERIES DE TAYLOR CON X0 = 1

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 1000)
y = np.cos(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.title('Gráfica de la función coseno')
plt.grid()
plt.show()
```

### SERIE DE TAYLOR DE ORDEN 0

```python
taylor_0 = np.cos(1) * np.ones_like(x)
plt.plot(x, y, label='cos(x)')
plt.plot(x, taylor_0, label='Taylor orden 0', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función coseno y su aproximación de Taylor en x0=1')
plt.axvline(x=1, color='red', linestyle='--', label='Punto de expansión x=1')
plt.grid()
plt.legend()
plt.show()
```

### SERIE DE TAYLOR DE ORDEN 0 Y 1

```python
taylor_1 = np.cos(1) - np.sin(1) * (x-1)
plt.plot(x, y, label='cos(x)')
plt.plot(x, taylor_0, label='Taylor orden 0', linestyle='--')
plt.plot(x, taylor_1, label='Taylor orden 1', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función coseno y sus aproximaciones de Taylor en x0=1')
plt.axvline(x=1, color='red', linestyle='--', label='Punto de expansión x=1')
plt.grid()
plt.legend()   
plt.show()
```

### SERIE DE TAYLOR DE ORDEN 0, 1 Y 2

```python
taylor_2 = np.cos(1) - np.sin(1) * (x-1) - (np.cos(1)/2) * (x-1)**2
plt.plot(x, y, label='cos(x)')
plt.plot(x, taylor_0, label='Taylor orden 0', linestyle='--')
plt.plot(x, taylor_1, label='Taylor orden 1', linestyle='--')
plt.plot(x, taylor_2, label='Taylor orden 2', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función coseno y su aproximación de Taylor en x0=1')
plt.axvline(x=1, color='red', linestyle='--', label='Punto de expansión x=1')
plt.grid()
plt.legend()
plt.show()
```

### SERIES DE TAYLOR DE ORDEN 0, 1, 2, 3 Y 4

```python
taylor_3 = np.cos(1) - np.sin(1) * (x-1) - (np.cos(1)/2) * (x-1)**2 + (np.sin(1)/6) * (x-1)**3
taylor_4 = np.cos(1) - np.sin(1) * (x-1) - (np.cos(1)/2) * (x-1)**2 + (np.sin(1)/6) * (x-1)**3 - (np.cos(1)/24) * (x-1)**4
plt.plot(x, y, label='cos(x)')
plt.plot(x, taylor_0, label='Taylor orden 0', linestyle='--')
plt.plot(x, taylor_1, label='Taylor orden 1', linestyle='--')
plt.plot(x, taylor_2, label='Taylor orden 2', linestyle='--')
plt.plot(x, taylor_3, label='Taylor orden 3', linestyle='--')
plt.plot(x, taylor_4, label='Taylor orden 4', linestyle='--')
plt.xlabel('x')
plt.axvline(x=1, color='red', linestyle='--', label='Punto de expansión x=1')
plt.ylabel('y')
plt.title('Gráfica de la función coseno y sus aproximaciones de Taylor en x0=1')
plt.grid()
plt.legend()
plt.show()
```

## Graficar la función 1/(1-x) con X0= 0.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 1000)
y = 1 / (1 - x)

mascara = np.abs(x - 1) > 0.01
x_clean = x[mascara]
y_clean = y[mascara]

plt.figure(figsize=(8, 5))
plt.plot(x_clean, y_clean, 'b', linewidth=1.5)
plt.axvline(x=1, color='r', linestyle='--', linewidth=1, label='Asíntota x=1')
plt.axvline(x=0, color='red', linestyle='--', label='Punto de expansión x=0')
plt.axhline(y=0, color='gray', linestyle='-', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('1/(1-x)')
plt.title('Gráfica de la función 1/(1-x)')
plt.xlim(-2, 2)
plt.ylim(-10, 10)   
plt.legend()
plt.grid(True)
plt.show()
```

### SERIES DE TAYLOR PARA ORDEN 0, 1, 2, 3 Y 4

```python
# Incluir la aproximación de orden 0, 1, 2, 3 y 4 de la función 1/(1-x) alrededor de x0=0
taylor_0 = 1 * np.ones_like(x)
taylor_1 = 1 + x
taylor_2 = 1 + x + x**2
taylor_3 = 1 + x + x**2 + x**3
taylor_4 = 1 + x + x**2 + x**3 + x**4
plt.figure(figsize=(8, 5))
plt.plot(x_clean, y_clean, 'b', linewidth=1.5, label='1/(1-x)')
plt.plot(x, taylor_0, 'r--', label='Taylor orden 0')
plt.plot(x, taylor_1, 'g--', label='Taylor orden 1')
plt.plot(x, taylor_2, 'm--', label='Taylor orden 2')
plt.plot(x, taylor_3, 'c--', label='Taylor orden 3')
plt.plot(x, taylor_4, 'y--', label='Taylor orden 4')
plt.axvline(x=1, color='r', linestyle='--', linewidth=1, label='Asíntota x=1')
plt.axhline(y=0, color='gray', linestyle='-', linewidth=0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función 1/(1-x) y sus aproximaciones de Taylor en x0=0')
plt.axvline(x=0, color='red', linestyle='--', label='Punto de expansión x=0')
plt.xlim(-2, 2)
plt.ylim(-10, 10)
plt.legend()
plt.grid(True)
plt.show()
```

## Graficar la función ln(x), x0=1

```python
# Graficar la función ln(x)
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0.01, 5, 1000)
y = np.log(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('ln(x)')
plt.title('Gráfica de la función ln(x)')
plt.axvline(x=1, color='red', linestyle='--', label='Punto de expansión x=1')
plt.xlim(-3, 4)
plt.ylim(-3, 3)
plt.grid()
plt.show()
```

### SERIES DE TAYLOR DE ORDEN 0, 1, 2, 3 Y 4 EN X0=1

```python
taylor_0 = np.log(1) * np.ones_like(x)
taylor_1 = np.log(1) + (1/1) * (x-1)
taylor_2 = np.log(1) + (1/1) * (x-1) - (1/2) * (x-1)**2
taylor_3 = np.log(1) + (1/1) * (x-1) - (1/2) * (x-1)**2 + (1/3) * (x-1)**3
taylor_4 = np.log(1) + (1/1) * (x-1) - (1/2) * (x-1)**2 + (1/3) * (x-1)**3 - (1/4) * (x-1)**4
plt.plot(x, y, label='ln(x)')  
plt.plot(x, taylor_0, label='Taylor orden 0', linestyle='--')
plt.plot(x, taylor_1, label='Taylor orden 1', linestyle='--')
plt.plot(x, taylor_2, label='Taylor orden 2', linestyle='--')
plt.plot(x, taylor_3, label='Taylor orden 3', linestyle='--')
plt.plot(x, taylor_4, label='Taylor orden 4', linestyle='--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfica de la función ln(x) y sus aproximaciones de Taylor en x0=1')
plt.axvline(x=1, color='red', linestyle='--', label='Punto de expansión x=1')
plt.xlim(-3, 4)
plt.ylim(-3, 3)
plt.legend()
plt.grid()
plt.show()
```

# Encuentre el polinomio de Lagrange para los siguientes datos y grafique

### 1. (0,0),(30,0.5),(60,√3/2),(90,1)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 30, 60, 90])
y = np.array([0, 0.5, np.sqrt(3)/2, 1])

plt.figure(figsize=(8,5))
plt.scatter(x, y, s=80)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Conjunto de puntos")
plt.show()
```

#### POLINOMIO DE LAGRANGE

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.Symbol('x')

x0, x1, x2, x3 = 0, 30, 60, 90
y0, y1, y2, y3 = 0, sp.Rational(1,2), sp.sqrt(3)/2, 1

L0 = ((x-x1)*(x-x2)*(x-x3))/((x0-x1)*(x0-x2)*(x0-x3))
L1 = ((x-x0)*(x-x2)*(x-x3))/((x1-x0)*(x1-x2)*(x1-x3))
L2 = ((x-x0)*(x-x1)*(x-x3))/((x2-x0)*(x2-x1)*(x2-x3))
L3 = ((x-x0)*(x-x1)*(x-x2))/((x3-x0)*(x3-x1)*(x3-x2))

P = sp.expand(y0*L0 + y1*L1 + y2*L2 + y3*L3)

print("L0(x) =", sp.simplify(L0))
print("L1(x) =", sp.simplify(L1))
print("L2(x) =", sp.simplify(L2))
print("L3(x) =", sp.simplify(L3))
print("\nP(x) =", P)

f = sp.lambdify(x, P, 'numpy')

xp = np.linspace(0, 90, 500)
yp = f(xp)

xd = np.array([0, 30, 60, 90])
yd = np.array([0, 0.5, np.sqrt(3)/2, 1])

plt.figure(figsize=(8,5))
plt.plot(xp, yp)
plt.scatter(xd, yd, s=80)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Polinomio de Lagrange")
plt.show()
```
